```python
import numpy as np
import pandas as pd
from Gen_data import SimulationStudy
import matplotlib.pyplot as plt
```

```python
sim1: SimulationStudy = SimulationStudy(p=20, mean_correlation=0.1, cor_variance=0.2, n=500, no_feat_cate=4, non_linear='linear')
simulation1 = sim1.create_dataset()
```

```python
plt.style.use('seaborn-v0_8')
y = simulation1['CATE']

fig = plt.figure(figsize=(10,10))
gs = fig.add_gridspec(nrows=5, ncols=4, hspace=0.4)
axs = gs.subplots(sharex=True, sharey=True)

feat = 0
for j in range(0, 5):
    for i in range(0, 4):
        feature = f'X{feat}'
        x = simulation1[feature]
        
        if feat < sim1.__getattribute__('no_feat_cate'):
            axs[j, i].scatter(x, y, s=5, marker='o', color = 'red')
            axs[j, i].set_title(feature)

        else:
            axs[j, i].scatter(x, y, s=5, marker='o', color = 'blue')
            axs[j, i].set_title(feature)

        if feat < 19:
            feat += 1
        else:
            feat = feat

fig.suptitle('Influence of Individual Features on the CATE function', y=0.96)
fig.text(0.5, 0.04, 'Feature', ha='center')  # Adjust vertical position
fig.text(0.04, 0.5, 'CATE', va='center', rotation='vertical')  # Adjust horizontal position
#axs[0, 1].plot(x, y, 'tab:orange')
#axs[0, 2].plot(x, y, 'tab:orange')
#axs[0, 1].set_title('Axis [0, 1]')
#axs[1, 0].plot(x, -y, 'tab:green')
#axs[1, 0].set_title('Axis [1, 0]')
#axs[1, 1].plot(x, -y, 'tab:red')
#axs[1, 1].set_title('Axis [1, 1]')

#.set(xlabel='Feature', ylabel='CATE')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```
