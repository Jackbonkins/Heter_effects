```python
from Gen_data import SimulationStudy
import matplotlib.pyplot as plt
```

```python
def plot_cate_feat_val(simulation):
    plt.style.use('seaborn-v0_8')

    y = simulation['CATE']
    feat_no = sim.__getattribute__('p')

    fig = plt.figure(figsize=(10,10))
    gs = fig.add_gridspec(nrows=5, ncols=4, hspace=0.4)
    axs = gs.subplots(sharex=True, sharey=True)

    feat = 0
    for j in range(0, 5): 
        for i in range(0, 4):
            feature = f'X{feat}'
            x = simulation[feature]
                       
            if feat < sim.__getattribute__('no_feat_cate'):
                axs[j, i].scatter(x, y, s=5, marker='o', color = 'red')
                axs[j, i].set_title(f'$X_{{{feat}}}$')

            else:
                axs[j, i].scatter(x, y, s=5, marker='o', color = 'blue')
                axs[j, i].set_title(f'$X_{{{feat}}}$')

            if feat < feat_no-1:
                feat += 1
            else:
                feat = feat

    #fig.suptitle('Influence of Individual Feature Values on the CATE Function', y=0.96)
    fig.text(0.5, 0.04, 'Feature Values', ha='center')  # Adjust vertical position
    fig.text(0.04, 0.5, 'CATE', va='center', rotation='vertical')  # Adjust horizontal position

    # Hide x labels and tick labels for top plots and y ticks for right plots.
    for ax in axs.flat:
        ax.label_outer()

```

```python
sim: SimulationStudy = SimulationStudy(p=20, mean_correlation=0.1, cor_variance=0.2, n=500, no_feat_cate=4, non_linear='linear')
simulation_linear = sim.create_dataset()
plot_cate_feat_val(simulation_linear)
```

```python
sim: SimulationStudy = SimulationStudy(p=20, mean_correlation=0.1, cor_variance=0.2, n=500, no_feat_cate=4, non_linear='quadratic')
simulation_quadratic = sim.create_dataset()
plot_cate_feat_val(simulation_quadratic)
```
